```python
df = df.dropna(subset=['salary'])  # kolom target harus ada
job_cols = ['jobType/0', 'jobType/1', 'jobType/2', 'jobType/3']
df[job_cols] = df[job_cols].fillna(False)
df['externalApplyLink'] = df['externalApplyLink'].fillna('Unavailable')
```

```python
df = df.drop_duplicates(subset=['company', 'positionName', 'location', 'salary', 'description'])
```

```python
df['salary_cleaned'] = df['salary'].astype(str).str.replace(r'[^\d.]', '', regex=True)
df['salary_num'] = pd.to_numeric(df['salary_cleaned'], errors='coerce')
```

```python
# PERBAIKAN 1: Membuat fitur desc_len dan posname_len sebelum discaling
df['desc_len'] = df['description'].astype(str).apply(len)
df['posname_len'] = df['positionName'].astype(str).apply(len)
```

```python
# Membuat binning untuk gaji
bins_salary = [0, 70000, 120000, 200000, df['salary_num'].max()]
labels_salary = ['Low', 'Medium', 'High', 'Very High']
df['salary_bin'] = pd.cut(df['salary_num'], bins=bins_salary, labels=labels_salary)
```

```python
# Membuat binning untuk rating
bins_rating = [0, 2, 3.5, 4.5, 5]
labels_rating = ['Poor', 'Average', 'Good', 'Excellent']
df['rating_bin'] = pd.cut(df['rating'], bins=bins_rating, labels=labels_rating)
```

```python
# Filter rating dan hapus baris yang target gajinya kosong
df = df[df['rating_bin'].isin(['Good', 'Excellent'])].copy()
df = df.dropna(subset=['salary_bin'])
```

```python
print("Distribusi kelas target:")
print(df['salary_bin'].value_counts())
print("-" * 40)
```

```python
# --- 2. SPLITTING DATA ---
# Menentukan target (y)
y = df['salary_bin']

# Definisikan daftar kolom kategori yang mau di-encode nanti
cols_to_encode = [
    'company', 'location', 'positionName',
    'jobType/0', 'jobType/1', 'jobType/2', 'jobType/3',
    'searchInput/country', 'searchInput/position'
]
cols_numeric = ['rating', 'desc_len', 'posname_len']

# PERBAIKAN: Masukkan gabungan kolom angka DAN kolom kategori ke dalam X
X = df[cols_numeric + cols_to_encode].copy()

# Train-test split (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# --- 3. CATEGORICAL ENCODING (ONE-HOT ENCODING) ---

# Baru jalankan proses get_dummies (Sekarang dijamin aman dari KeyError!)
X_train = pd.get_dummies(X_train, columns=cols_to_encode, drop_first=True)
X_test = pd.get_dummies(X_test, columns=cols_to_encode, drop_first=True)

# Antisipasi perbedaan kolom kategori unik antara data train dan test
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

print("Encoding Berhasil!")
```

```python
# --- 4. FEATURE SCALING (STANDARD_SCALER) ---
# PERBAIKAN 3: Hanya menskalakan fitur numerik independen yang ada di X
cols_to_scale = ['rating', 'desc_len', 'posname_len']

scaler = StandardScaler()
X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])

print("Proses Berhasil! Data siap dimasukkan ke dalam Model Machine Learning.")
```

```python
# --- 5. EKSPOR DATA HASIL PREPROCESSING ---

# 1. Salin data X_train yang sudah melalui proses Encoding & Scaling
Xy_train = X_train.copy()

# 2. Gabungkan kembali target (salary_bin) ke dalam dataframe tersebut
Xy_train['salary_bin'] = y_train

# 3. Definisikan nama folder di satu variabel agar pasti sama
folder_name = "preprocessing"

# 4. Buat folder baru jika belum ada
os.makedirs(folder_name, exist_ok=True)

# 5. Tentukan alamat penyimpanan file CSV menggunakan variabel folder_name
output_path = os.path.join(folder_name, "JobDataset_preprocessed.csv")

# 6. Ekspor dataframe menjadi file CSV tanpa menyertakan indeks baris
Xy_train.to_csv(output_path, index=False)

print("✅ File disimpan di:", output_path)
```
